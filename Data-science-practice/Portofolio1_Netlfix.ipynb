```python
import pandas as pd
netflix_raw = pd.read_csv("netflix_titles.csv")
print(netflix_raw.columns)
```

```python
netflix_raw = pd.DataFrame(netflix_raw)
print(netflix_raw.head())
```

```python
print(netflix_raw['country'].unique())
```

```python
netflix_raw.drop(['date_added', 'director', 'cast', 'show_id', 'description'], axis=1, inplace=True)
print(netflix_raw.head())
```

```python
print(netflix_raw['release_year'].min())
print(netflix_raw['release_year'].max())
```

```python
print("US")
print((netflix_raw['country']=='United States').value_counts())
print("Turkey")
print((netflix_raw['country']=='Turkey').value_counts())
print("Mexico")
print((netflix_raw['country']=='Mexico').value_counts())
print("Top 10 COUNTRY")
top_country = netflix_raw['country'].value_counts()[:21]
print(top_country)
```

```python
import matplotlib.pyplot as plt
data = netflix_raw['country'].value_counts()[10:20]
plt.plot(data)
plt.xticks(rotation=60)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,6))
netflix_raw['country'].sort_values(ascending=True).value_counts()[:10].plot(kind='barh')
plt.xticks(rotation=60)
plt.title("Top 10 Country Film Maker - Netflix", fontsize=18)
plt.text(1200,9, 'Data source: Kaggle')

plt.xlabel('Amount of Film')
plt.savefig('top10countryfilmmaker.jpg', bbox_inches='tight')
plt.show()
```

```python
top_netflix = pd.DataFrame(netflix_raw['country'].value_counts().reset_index().values,
             columns=["country", "films"])
top_netlfix = top_netflix.sort_index(axis=0, ascending=True)
def find(data):
    i =0
    status = False
    for x in top_netflix['country']:
        if x == data:
            index = i
            status = True
            break
        i += 1
    if status == True :
        print("Top rank -", index+1)
    else :
        print("404!")
find("Indonesia")
```

```python
netflix_raw['type'].unique()
netflix_raw.drop(netflix_raw[netflix_raw.type == "TV Show"].index, inplace=True)
netflix_raw['type'].unique()
```

```python
print(netflix_raw['duration'].dtype)
netflix_raw['duration'] = netflix_raw['duration'].str.split(' ', 1, expand=True)
print(netflix_raw.head())
print(netflix_raw['duration'].dtype)
netflix_raw['duration'].astype(int)

```

```python
netflix_raw['duration'] = pd.to_numeric(netflix_raw['duration'])
print(netflix_raw['duration'].dtype)
netflix_raw.drop(['mins'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
netflix_raw['type'] = LE.fit_transform(netflix_raw['type'])
netflix_raw['country'] = LE.fit_transform(netflix_raw['country'])
netflix_raw['release_year'] = LE.fit_transform(netflix_raw['release_year'])
netflix_raw['rating'] = LE.fit_transform(netflix_raw['rating'])
netflix_raw['duration'] = LE.fit_transform(netflix_raw['duration'])
```
